```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('E:\Gold-Price-Forecast-Based-on-LSTM-Model\dataset\data_gold_1.csv')
df.head()
```

```python
# datetime
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
df = df.sort_values(by='date', ascending=True, ignore_index=True)
df.head()
```

```python
# Visualization
plt.figure(figsize=(15, 7))
sns.lineplot(data=df, x='date', y='price_gold', color='gold')
plt.show()
```
